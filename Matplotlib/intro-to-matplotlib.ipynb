```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales_price = car_sales.Price.str.replace("$", "").str.replace(",", "").astype(float).astype(int)
car_sales_price
```

```python
car_sales.Price = car_sales_price
car_sales
```

```python
car_sales["Total Sales"] = car_sales.Price.cumsum()
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2024", periods=len(car_sales))
car_sales
```

```python
#lets plot total sales
car_sales.plot(x="Sale Date", y="Total Sales");
plt.show()
```

```python
car_sales.plot(x="Odometer (KM)", y = "Price", kind="scatter")
plt.show()
```

```python
car_sales.plot.bar()
plt.show()
```

```python
car_sales.plot(kind='bar')
plt.show()
```

```python
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar")
plt.show()
```

```python
car_sales.plot.hist()
plt.show()
```

```python
car_sales["Price"].plot.hist()
plt.show()
```

```python
car_sales["Odometer (KM)"].plot.hist()
plt.show()
```

```python
car_sales.plot.hist(figsize=(10,30), subplots=True)
plt.show()
```

```python
## OO method
fig, ax = plt.subplots(figsize = (10,6))
#plot the data
scatter = ax.scatter(x=car_sales.Price,
                     y = car_sales["Odometer (KM)"],
                     c = car_sales["Doors"])
#customize the plot
ax.set(title= "Car price and Odometer",
    xlabel = "Price",
    ylabel = "Odometer reading")

#Add a legend
ax.legend(*scatter.legend_elements(), title="Doors")

#Add a horizontal line and vertical line on mean
ax.axhline(car_sales["Odometer (KM)"].mean(), linestyle="--")
ax.axvline(car_sales["Price"].mean(), linestyle=":")

plt.show()
                     
```
